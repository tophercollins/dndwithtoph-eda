### 1.2 Remove Useless Data/Features

* payout_id, paid_period, payout_date, status - All details for issues with payment but not neccesary for analysis
* owner_name, leader_user_id, leader_name- Details related to the teacher
* section_link, loaded_at, timezone - not useful data
* duplicate_number - unclear what refers to so dropping

```python
# Drop those columns, we'll use inplace=True overwriting those columns inplace.

dndwithtoph.drop(['payout_id','paid_period','payout_date','status',
           'owner_name','leader_user_id','leader_name','section_link',
            'loaded_at','duplicate_number','timezone'], axis=1, inplace=True)

dndwithtoph.head()
```

### 1.3 Convert Appropriate Data to Numerical

payout_dollars is not in numerical format so let's adjust

```python
dndwithtoph["payout_dollars"] =  dndwithtoph["payout_dollars"].str.replace('[\$\.]','',regex=True).astype(int)

dndwithtoph.head()
```

```python
# Values are too high now must divide by 100 to account for the 00s aftet the decimal point in original string

dndwithtoph.loc[:,'payout_dollars'] /= 100

dndwithtoph.head()
```

### 1.4 Rename Columns

```python
dndwithtoph.rename(columns={'class_name':'Class',
                            'first_meeting_of_week_date':'Date & Time',
                            'meetings':'Meetings',
                            'hours_taught':'Hours',
                           'payout_dollars':'Payout ($)',
                           'num_enrollments':'Students'},inplace=True)

dndwithtoph.head()
```

### 2.2 Summarise Data
* On-going adventures can be studied against each other to get a better scope of each other
* We can make a fresh adventures dataframe from the data combined from the sum and mean groupbys.

```python
# Let's make a new adventure DataFrame from the summaries here
    
adventures = dndwithtoph.groupby('Class').mean(numeric_only=True)

# Rename the Columns

adventures.rename(columns={'Hours':'Mean Hours',
                           'Payout ($)':'Mean Payout ($)',
                           'Students':'Mean Students',
                           'Payout($) per Student':'Mean Payout ($) per Student'},
                 inplace=True)

adventures
```

```python
# Add the Total Hours and Total Payout from the sums groupby

adventures['Total Hours'] = dndwithtoph.groupby('Class').sum(numeric_only=True).loc[:,'Hours']
adventures['Total Payout ($)'] = dndwithtoph.groupby('Class').sum(numeric_only=True).loc[:,'Payout ($)']

adventures
```

```python
# Drop the non-adventure classes

adventures.drop(["DnD with Toph - Dungeons & Dragons Social Club",
                 "Let's Play Chess! Weekly Tournaments - Suitable For All Skill Levels!",
                "DnD with Toph - Special One-off Class!",
                "Drama Class - Theatre Games & Exercises for Kids",
                "Dungeons and Dragons (D&D) Catch Up Class",
                "Dungeons and Dragons (D&D) Introduction Part 1: Creating a Character!"],
               inplace=True)

# It's not clear Mean Hours also doesn't add anything as it's the same for all classes

adventures.drop('Mean Hours',axis=1,inplace=True)

adventures
```

```python
# Excessive decimal places so let's round everything down

adventures = adventures.round(2)

adventures
```

```python
# Let's add payout per hours as another column to see time efficiency

adventures['Payout ($) per Hour'] = adventures.loc[:,'Total Payout ($)']/adventures.loc[:,'Total Hours']

```

```python
adventures = adventures.round(2)

adventures
```

```python
adventures.sort_values('Payout ($) per Hour',ascending=False)
```
